```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('DistanceHealth.csv')
data.head(-1)
```

```python
average=np.mean(data.eyes)
print(average)
data=data[(data.eyes < average*1.05)&(data.eyes > average*0.95)]
data.head(-1)
```

```python
data.to_excel('test1.xlsx',index=0)
data.to_csv('test1.csv',index=0)
```

```python
#讀取xlsx
import openpyxl
from openpyxl import load_workbook
wb = load_workbook('test1.xlsx')
print(wb.sheetnames)
sheet = wb.active
```

```python
frame = []
for i in range(2, len(data)):
    frame.append(sheet.cell(row=i, column=1).value)
print(frame)
```
