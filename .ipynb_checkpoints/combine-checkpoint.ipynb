```python
import pandas as pd
df = pd.read_csv('SW_line_Distance_health.csv')
df1 = pd.read_csv('SW_line_Distance_test3.csv')
df2 = pd.read_csv('SW_line_Distance_test2.csv')
df3 = pd.read_csv('SW_line_Distance_test1.csv')
df4 = pd.read_csv('SW_line_Distance_pain.csv')
print(df)
```

```python
#怎麼新增column並加上資料
#df.insert(11,column="Pain level",value=1)
#刪除資料
#df.drop(labels="frame", axis=1)
def csv2(x,label):
    x.insert(11,column="PainLevel",value=label)
    x=x.drop(labels="frame", axis=1)
    print(x)
    return x

df=csv2(df,0)
df4=csv2(df4,1)
df1=df1.drop(labels="frame", axis=1)
print(df1)
df2=df2.drop(labels="frame", axis=1)
print(df2)
df3=df3.drop(labels="frame", axis=1)
print(df3)
```

```python
#合併CSV與匯出
c = pd.concat([df,df4])
c.to_csv('SW_line_train.csv',index=0)

c = pd.concat([df1,df2,df3])
c.to_csv('SW_line_test.csv',index=0)
```
