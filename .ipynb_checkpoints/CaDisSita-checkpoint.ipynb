```python
import openpyxl
import os
import shutil #python copy model
import numpy as np
import math
import csv
import pandas as pd
from openpyxl import load_workbook
```

```python
# 計算兩點距離
def eucliDist(A,B):
    return math.sqrt(sum([(a - b)**2 for (a,b) in zip(A,B)]))
```

```python
#計算角度
def sita(a,b,c):
    if(a+b>c and a+c>b and b+c>a):    #判断三边是否能构成三角形
         A = math.degrees(math.acos((a*a-b*b-c*c)/(-2*b*c)))     #求角A
         B = math.degrees(math.acos((b*b-a*a-c*c)/(-2*a*c)))  #求角B
         C = math.degrees(math.acos((c*c-a*a-b*b)/(-2*a*b)))  #求角C
         return(A,B,C)      #输出角A #输出角B #输出角C
    else:
        #print("三边无法构成三角形")
        return(0,0,0)
```

```python
def CAandDis(fileNameIn,fileNameOut_CA,fileNameOut_Dis):
    #輸入路徑!!!!!!!!!!
    wb = load_workbook(fileNameIn)
    sheet = wb.active
    
    #整行資料儲存轉成list
    X_data = []
    for cell in list(sheet.columns)[13]:
        X_data.append(cell.value)

    Y_data = []
    for cell in list(sheet.columns)[14]:
        Y_data.append(cell.value)

    frame=[]
    for cell in list(sheet.columns)[9]:
        frame.append(cell.value)
    
    #建立CA.csv
    with open(fileNameOut_CA, 'w', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(['frame', 'ReyeX', 'ReyeY', 'LeyeX', 'LeyeY', 'RearX', 'RearY', 'LearX', 'LearY', 'noseX', 'noseY'])

    Num=int((len(X_data)-1)/5)
    for i in range(int(Num)):
        #未旋轉 
        (x1, y1)=(X_data[i+1], Y_data[i+1])
        (x2, y2)=(X_data[i+Num+1], Y_data[i+Num+1])
        (x3, y3)=(X_data[i+2*Num+1], Y_data[i+2*Num+1])
        (x4, y4)=(X_data[i+3*Num+1], Y_data[i+3*Num+1])
        (x5, y5)=(X_data[i+4*Num+1], Y_data[i+4*Num+1])
        
        #寫入CA.csv
        with open(fileNameOut_CA, 'a+', newline = '') as csvFile:
            csvWriter = csv.writer(csvFile)
            csvWriter.writerow([frame[i+1], x1, y1, x2, y2, x3, y3, x4, y4, x5, y5])
    
    #讀入CA.csv
    CA=pd.read_csv(fileNameOut_CA).values
    
    #建立Dis.csv
    with open(fileNameOut_Dis, 'w', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(['frame', 'eyes', 'eyeRearR', 'eyeRearL', 'eyeRnose', 'eyeLearR', 'eyeLearL', 'eyeLnose', 'ears'])  

    #計算歐式距離
    for i in CA:
        Q1=i[1:3];Q2=i[3:5];Q3=i[5:7];Q4=i[7:9];Q5=i[9:11];
        d=eucliDist(Q1,Q2);
        d1=eucliDist(Q1,Q3);d2=eucliDist(Q1,Q4);d3=eucliDist(Q1,Q5);
        d4=eucliDist(Q2,Q3);d5=eucliDist(Q2,Q4);d6=eucliDist(Q2,Q5);
        d7=eucliDist(Q3,Q4);
        
        #寫入Distance.csv
        with open(fileNameOut_Dis, 'a+', newline = '') as csvFile:
            csvWriter = csv.writer(csvFile)
            csvWriter.writerow([i[0] ,d, d1, d2, d3, d4, d5, d6, d7])    
```

```python
def sitaFace(fileNameIn,fileNameOut):
    #要輸入的檔名!!!!!!!!!
    D=pd.read_csv(fileNameIn).values

    #輸出檔名.csv!!!!!
    file = fileNameOut
    with open(file, 'w', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(['frame','sita1','sita2','sita3','sita4','sita5',
                            'sita6','sita7','sita8','sita9','sita10',
                            'sita11','sita12','sita13','sita14','sita15'])
    #計算角度
    for i in D:
        s=sita(i[1],i[2],i[5]);s1=sita(i[1],i[3],i[6]);s2=sita(i[1],i[4],i[7]);
        o=sita(i[8],i[2],i[3]);o1=sita(i[8],i[5],i[6]);
       
        with open(file, 'a+', newline = '') as csvFile:
            csvWriter = csv.writer(csvFile)
            csvWriter.writerow([i[0], s[0], s[1], s[2], s1[0], s1[1], s1[2], s2[0], s2[1], s2[2],
                                o[0], o[1], o[2], o1[0], o1[1], o1[2]
                               ])     
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn.utils import shuffle
from sklearn.model_selection import cross_val_score
```

```python
dfh=pd.read_csv('DistanceHealth.csv')
dfh.insert(9,column="PainLevel",value=0)
dfh=dfh.drop(labels=['frame'],axis=1)

dfp=pd.read_csv('DistancePain.csv')
dfp.insert(9,column="PainLevel",value=1)
dfp=dfp.drop(labels=['frame'],axis=1)


#Num of pain = Num of health

if len(dfp) >= len(dfh):
    dfp=pd.DataFrame.sample(dfp,n=len(dfh),random_state=1).reset_index(drop=True)
    print(len(dfp))
elif len(dfp) < len(dfh):
    dfh=pd.DataFrame.sample(dfh,n=len(dfp),random_state=1).reset_index(drop=True)
    print(len(dfh))
    
# df of train = health+pain
df = shuffle(pd.concat([dfh,dfp]),random_state=1).reset_index(drop=True)

data = df.drop(labels=['PainLevel'],axis=1).values
target = df.PainLevel.values
```

```python
ss = StandardScaler()
data_Standar = ss.fit_transform(data)

from sklearn.model_selection import RandomizedSearchCV
svc = svm.SVC(kernel='linear')
param_dist = {
        'C':np.linspace(0.1,1000,10000)
        }

grid = RandomizedSearchCV(svc,param_dist,cv = 3,scoring = 'accuracy',n_iter=100,n_jobs = -1,verbose=1)

grid.fit(data_Standar,target)
best_estimator = grid.best_estimator_
print(best_estimator)
print(grid.best_score_)
```

```python
ss = StandardScaler()
data_Standar = ss.fit_transform(data)

c_range = [0.001,0.003,0.01,0.03,0.1,0.3,1,3,10,30,100,300,1000]
c_scores = []

for c_number in c_range:
    svc = svm.SVC(C=c_number, kernel='linear',probability=True)
    scores = cross_val_score(svc,data_Standar,target,cv=3,scoring='accuracy',verbose=1,n_jobs=-1)
    c_scores.append(scores.mean())

plt.plot(c_range,c_scores)
plt.xlabel('Value of c for SVM')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```
