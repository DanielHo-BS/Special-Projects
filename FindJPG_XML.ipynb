```python
import openpyxl
import os
import shutil #python copy model
path_jpg = 'D:/rat_video/20191226_CFA_ipl_pain_face/CFA_D1_diclofenac_treatment/3hr/10mg/KT/result' #jpg文件路径
copy_to = 'D:/rat_video/yolo_facial_pattern_data/20191226_CFA/20191226_CFA_KT/KT123/test3' #拷贝保存路径
```

```python
#讀取xlsx
from openpyxl import load_workbook
wb = load_workbook('D:/rat_video/yolo_facial_pattern_data/20191226_CFA/20191226_CFA_KT/KT123/test3.xlsx')
print(wb.sheetnames)
```

```python
sheet = wb.active
```

```python
#input num of frame
Num=9725
dReye = []
for i in range(2, Num):
    dReye.append(sheet.cell(row=i, column=10).value)
print(dReye)
#data_Reye to list
```
