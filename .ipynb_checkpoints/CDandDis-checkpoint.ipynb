```python
import openpyxl
import os
import shutil #python copy model
import numpy as np
import math
import csv
```

```python
#輸入路徑!!!!!!!!!!
from openpyxl import load_workbook
wb = load_workbook('test3.xlsx')
print(wb.sheetnames)
```

```python
sheet = wb.active

frame=[]
for cell in list(sheet.columns)[9]:
    frame.append(cell.value)
print(frame)
```

```python
#整行資料儲存轉成list
X_data = []
for cell in list(sheet.columns)[13]:
    X_data.append(cell.value)
#print(X_data)

Y_data = []
for cell in list(sheet.columns)[14]:
    Y_data.append(cell.value)
#print(Y_data)
```

```python
#要輸出的檔名!!!!!!!!!
file = 'SW_line_CoordinateAlignment_test3.csv' 
with open(file, 'w', newline = '') as csvFile:
    csvWriter = csv.writer(csvFile)
    csvWriter.writerow(['frame', 'ReyeX', 'ReyeY', 'LeyeX', 'LeyeY', 'RearX', 'RearY', 'LearX', 'LearY', 'noseX', 'noseY'])

#設定多少的樣本
Num=int((len(X_data)-1)/5)
for i in range(int(Num)):
    #未旋轉 
    (x1, y1)=(X_data[i+1], Y_data[i+1])
    (x2, y2)=(X_data[i+Num+1], Y_data[i+Num+1])
    (x3, y3)=(X_data[i+2*Num+1], Y_data[i+2*Num+1])
    (x4, y4)=(X_data[i+3*Num+1], Y_data[i+3*Num+1])
    (x5, y5)=(X_data[i+4*Num+1], Y_data[i+4*Num+1])
    #旋轉角度
    dY=y1-y2
    dX=x1-x2
    angle = -np.degrees(np.arctan2(dY, dX))
    #print("angle is ",angle)
    #兩眼中心點
    center=np.array([0, 0])
    center[0]=(x1+x2)/2
    center[1]=(y1+y2)/2
    #print("center is ",center)
    #旋轉後
    tmp_angle= (angle*math.pi)/180
    (tmp_x1, tmp_y1)=(x1-center[0], y1-center[1])
    (tmp_x2, tmp_y2)=(x2-center[0], y2-center[1])
    (tmp_x3, tmp_y3)=(x3-center[0], y3-center[1])
    (tmp_x4, tmp_y4)=(x4-center[0], y4-center[1])
    (tmp_x5, tmp_y5)=(x5-center[0], y5-center[1])
    X1= int(tmp_x1*math.cos(tmp_angle)-tmp_y1*math.sin(tmp_angle)+center[0])
    Y1= int(tmp_x1*math.sin(tmp_angle)+tmp_y1*math.cos(tmp_angle)+center[1])
    X2= int(tmp_x2*math.cos(tmp_angle)-tmp_y2*math.sin(tmp_angle)+center[0])
    Y2= int(tmp_x2*math.sin(tmp_angle)+tmp_y2*math.cos(tmp_angle)+center[1])
    X3= int(tmp_x3*math.cos(tmp_angle)-tmp_y3*math.sin(tmp_angle)+center[0])
    Y3= int(tmp_x3*math.sin(tmp_angle)+tmp_y3*math.cos(tmp_angle)+center[1])
    X4= int(tmp_x4*math.cos(tmp_angle)-tmp_y4*math.sin(tmp_angle)+center[0])
    Y4= int(tmp_x4*math.sin(tmp_angle)+tmp_y4*math.cos(tmp_angle)+center[1])
    X5= int(tmp_x5*math.cos(tmp_angle)-tmp_y5*math.sin(tmp_angle)+center[0])
    Y5= int(tmp_x5*math.sin(tmp_angle)+tmp_y5*math.cos(tmp_angle)+center[1])
    #print("x1 ",X1, Y1)
    #print("x2 ",X2, Y2)
    #print("x3 ",X3, Y3)
    #print("x4 ",X4, Y4)
    #print("x5 ",X5, Y5)
    with open(file, 'a+', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow([frame[i+1], X1, Y1, X2, Y2, X3, Y3, X4, Y4, X5, Y5])
```

```python
# 計算兩點距離
def eucliDist(A,B):
    return math.sqrt(sum([(a - b)**2 for (a,b) in zip(A,B)]))
```

```python
#要輸入的檔名!!!!!!!!!
file = 'SW_line_CoordinateAlignment_test3.csv' 

#讀取原本座標
with open(file) as csvfile:
    rows = csv.reader(csvfile)
    
    # 以迴圈輸出每一列
    i=0
    X=np.zeros(11)
    tmp_X=np.zeros(11)
    for row in rows:
      print(row)
      if i>0:
        j=0
        while j<=10:
          tmp_X[j]=int(row[j])
          j += 1
        X=np.vstack((X,tmp_X))
      i+=1
X=np.delete(X, 0, axis=0)
```

```python
#要輸出的檔名 !!!!!!!
file = 'SW_line_Distance_test3.csv' 
with open(file, 'w', newline = '') as csvFile:
    csvWriter = csv.writer(csvFile)
    csvWriter.writerow(['frame', 'eyes', 'eyeRearR', 'eyeRearL', 'eyeRnose', 'eyeLearR', 'eyeLearL', 'eyeLnose', 'ears', 'earRnose', 'earLnose'])
    

#計算各點歐式距離
for i in X:
    Q1=i[1:3];Q2=i[3:5];Q3=i[5:7];Q4=i[7:9];Q5=i[9:11];
    d=eucliDist(Q1,Q2);
    d1=eucliDist(Q1,Q3);d2=eucliDist(Q1,Q4);d3=eucliDist(Q1,Q5);
    d4=eucliDist(Q2,Q3);d5=eucliDist(Q2,Q4);d6=eucliDist(Q2,Q5);
    d7=eucliDist(Q3,Q4);d8=eucliDist(Q3,Q5);
    d9=eucliDist(Q4,Q5);
    with open(file, 'a+', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow([i[0] ,d, d1, d2, d3, d4, d5, d6, d7, d8, d9]) 
```
