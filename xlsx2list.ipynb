```python
import openpyxl
import os
import shutil #python copy model
```

```python
from openpyxl import load_workbook
wb = load_workbook('health.xlsx')
print(wb.sheetnames)
```

```python
sheet = wb.active
j1 = sheet['J1']
print(f'({j1.column}, {j1.row}) is {j1.value}')
```

```python
for row in sheet.rows:
    for cell in row:
        print(cell.value)
#rows/columns可互換
#全部的儲存個格
```

```python
for cell in list(sheet.columns)[9]:
    print(cell.value)
```

```python
data = []
for cell in list(sheet.columns)[9]:
    data.append(cell.value)
print(data)
#整行資料儲存轉成list
```

```python
dReye = []
for i in range(2, 1459):
    dReye.append(sheet.cell(row=i, column=10).value)
print(dReye)
#data_Reye to list
```
