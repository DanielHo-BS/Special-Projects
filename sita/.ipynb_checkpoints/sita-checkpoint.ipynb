```python
import math
import os
import numpy as np
import math
import csv
import pandas as pd

def sita(a,b,c):
    if(a+b>c and a+c>b and b+c>a):    #判断三边是否能构成三角形
         A = math.degrees(math.acos((a*a-b*b-c*c)/(-2*b*c)))     #求角A
         B = math.degrees(math.acos((b*b-a*a-c*c)/(-2*a*c)))  #求角B
         C = math.degrees(math.acos((c*c-a*a-b*b)/(-2*a*b)))  #求角C
         return(A,B,C)      #输出角A #输出角B #输出角C
    else:
        #print("三边无法构成三角形")
        return(0,0,0)
```

```python
def sitaFace(fileNameIn,fileNameOut):
    #要輸入的檔名!!!!!!!!!
    file = fileNameIn

    #讀取原本座標
    with open(file) as csvfile:
        rows = csv.reader(csvfile)
    
        # 以迴圈輸出每一列
        i=0
        X=np.zeros(11)
        tmp_X=np.zeros(11)
        for row in rows:
          #print(row)
          if i>0:
            j=0
            while j<=10:
              tmp_X[j]=int(float(row[j]))
              j += 1
            X=np.vstack((X,tmp_X))
          i+=1
    X=np.delete(X, 0, axis=0)
    
    #計算角度
    #輸出檔名.csv!!!!!
    file = fileNameOut
    with open(file, 'w', newline = '') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(['frame','sita1','sita2','sita3','sita4','sita5','sita6','sita7','sita8','sita9','sita10',
                            'sita11','sita12','sita13','sita14','sita15','sita16','sita17','sita18','sita19','sita20',
                            'sita21','sita22','sita23','sita24','sita25','sita26','sita27','sita28','sita29','sita30'])
    for i in X:
        s=sita(i[1],i[2],i[5]);s1=sita(i[1],i[3],i[6]);s2=sita(i[1],i[4],i[7]);
        o=sita(i[8],i[2],i[3]);o1=sita(i[8],i[5],i[6]);o2=sita(i[8],i[9],i[10]);
        d=sita(i[2],i[4],i[9]);d1=sita(i[6],i[7],i[10]);d2=sita(i[3],i[4],i[10]);d3=sita(i[5],i[7],i[9]);
        with open(file, 'a+', newline = '') as csvFile:
            csvWriter = csv.writer(csvFile)
            csvWriter.writerow([i[0], s[0], s[1], s[2], s1[0], s1[1], s1[2], s2[0], s2[1], s2[2],
                                o[0], o[1], o[2], o1[0], o1[1], o1[2], o2[0], o2[1], o2[2],
                                d[0], d[1], d[2], d1[0], d1[1], d1[2], d2[0], d2[1], d2[2], d3[0], d3[1], d3[2] ]) 
    df=pd.read_csv(fileNameOut)
    df.head(-1)
```

```python
#輸入 距離特徵 & 輸出檔名
sitaFace('DistanceHealth.csv','KT_sitaHealth.csv' )
sitaFace('DistancePain.csv','KT_sitaPain.csv' )
sitaFace('Distance1.csv','KT_sita1.csv' )
sitaFace('Distance2.csv','KT_sita2.csv' )
sitaFace('Distance3.csv','KT_sita3.csv' )
```
