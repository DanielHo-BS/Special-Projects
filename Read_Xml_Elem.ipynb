```python
import xml.etree.ElementTree as ET

# 從檔案載入並解析 XML 資料
tree = ET.parse('health.xml')
root = tree.getroot()
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
for elem in tree.iter():
   print (elem.tag, elem.attrib)
```
